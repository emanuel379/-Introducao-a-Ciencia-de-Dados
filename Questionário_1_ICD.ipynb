**2 -** Leia os dados com erro novamente e insira a media de cada atributo onde for encontrada NaN e '?'. Quais os valores das medianas obtidas para os atributos?

```python
df = pd.read_csv('data/iris-with-errors.csv')
df
```

```python
# Substituindo '?' por NaN
df = df.replace('?', np.nan)
df
```

```python
# Substituindo valores NaN pela média

df['sepal_length'] = pd.to_numeric(df['sepal_length'], errors='coerce') # Convertendo valores para numéricos
df['sepal_length'] = df['sepal_length'].fillna(df['sepal_length'].mean())

df['sepal_width'] = pd.to_numeric(df['sepal_width'], errors='coerce')
df['sepal_width'] = df['sepal_width'].fillna(df['sepal_width'].mean())

df['petal_width'] = pd.to_numeric(df['petal_width'], errors='coerce')
df['petal_width'] = df['petal_width'].fillna(df['petal_width'].mean())
df
```

```python
# Encontrando as medianas
df.median()
```

**4 -** Considere a base de dados do Titanic, disponível na plataforma Kaggle. Calcule a correlação entre as variáveis. Quais são as três variáveis mais correlacionadas?

```python
import pandas as pd
import numpy as np
import requests
from io import StringIO

# URL da base de dados no Kaggle
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

# Fazer a requisição à URL e obter o conteúdo
response = requests.get(url)
data = response.text

# Criar um DataFrame a partir dos dados
df = pd.read_csv(StringIO(data))
df
```

```python
# Eliminando linhas e colunas consideradas desnecessárias

df.drop(['PassengerId','Name', 'Ticket', 'Cabin'],axis = 1, inplace = True)
df = pd.get_dummies(df)
df
```

```python
# Contabilizando Not a Number
nan_counts = df.isna().sum()

print(nan_counts)
```

```python
df = df.dropna()
df
```

```python
# Observando a correlação entre as variáveis

plt.figure(figsize = (10, 10))
sns.heatmap(df.corr(), annot = True)
plt.title('Correlação entre variáveis', fontsize=15)
plt.show()
```
